1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are recognized as being distinct from one another because they differ only by the use of upper or lower-case letters. For instance, "Love" is likely to occur at the beginning of a new sentence or as a proper name - the first letter being upper-case makes it a proper noun, whereas "love" is a regular word and should always be lower-case.

They will be treated as such by the model unless you normalize the text by converting it into all lower-case letters. Hence, why we convert text into lowercase characters.

2. What happens if stopwords are not removed?

By not removing stop words (such as is, the, and), you increase your vocabulary size without any benefit to your model (i.e., make your model slower and less effective), reduce the influence of important words relative to more common words, and lower the performance of your model overall.

3. Two cases where removing stopwords is harmful

**Sentiment Analysis**

Example: "I am not happy"

Removing "not" changes meaning → becomes positive

**Question Answering / Search Systems**

Example: "What is the capital of India?"

Removing "what" or "is" can distort intent

4. Stemming vs Lemmatization

Use Stemming: For large-scale text retrieval, search engines, or scenarios prioritizing speed over accuracy.
Approach: Rule-based cutting
Output: Not always real word

Use Lemmatization: For chatbots, text analysis, and machine learning models where identifying the exact root meaning is crucia
Approach:	Vocabulary + grammar
Output: Real meaningful word

```python
# TASK 2: Advanced Preprocessing Function
import re

def preprocess_text(text):
    if not text or text.strip() == "":
        return [], ""

    text = re.sub(r"http\S+|www\S+", "", text) # To Remove URLs

    text = re.sub(r"\S+@\S+", "", text) # To Remove emails

    text = re.sub(r"\d+", "", text)   # To Remove numbers

    text = text.lower() # For Lowercase

    # Handle repeated characters (soooo → soo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters (keep words)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenization
    tokens = text.split()

    # Remove short words (≤2) except "no", "not"
    cleaned_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]
    # Final cleaned sentence
    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

```python
# TASK 3: Stress Testing
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sentences:
    tokens, clean_sent = preprocess_text(s)
    print("\nOriginal:", s)
    print("Tokens:", tokens)
    print("Cleaned:", clean_sent)
```

```python
# TASK 4: Token Analytics
def analyze_tokens(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total if total > 0 else 0
    return total, unique, avg_len

for s in sentences:
    tokens, _ = preprocess_text(s)
    total, unique, avg = analyze_tokens(tokens)
    print(f"\nSentence: {s}")
    print("Total:", total, "Unique:", unique, "Avg Length:", avg)
```

Analysis Questions

Most meaningful tokens retained:

```python
# TASK 5: Frequency Analysis
from collections import Counter
all_tokens = []

for s in sentences:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

counter = Counter(all_tokens)
print("Top 10 frequent:", counter.most_common(10))
print("Least 5 frequent:", counter.most_common()[-5:])
```

```python
# TASK 6: Build Full Pipeline
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(sentences)
print(result)
```
